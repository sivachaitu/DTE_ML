```python
import numpy as np
from sklearn import preprocessing, model_selection, neighbors, svm
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_csv(r'C:\Users\sivac\Desktop\ML\archive\breast-cancer-wisconsin.data')
```

```python
df
```

```python
df.replace('?', -99999, inplace = True)
df.drop(['id'],1, inplace=True)  #if inplace = True is not used the changes are not being comitted hence affecting the outcome
```

```python
df.head()
```

```python
X = np.array(df.drop(['class'],axis= 1))
```

```python
y = np.array(df['class'])
```

```python
print(np.shape(y))
print(np.shape(X))
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.2)
```

```python
clf = neighbors.KNeighborsClassifier (n_jobs = -1) #imply change clf = svm.svc() to use svm instead of knn
clf.fit(X_train, y_train)
with open('kneighbor.pickle','wb') as f : #pickling .......
    pickle.dump(clf,f)
    
pickle_in = open('kneighbor.pickle','rb') 
clf = pickle.load(pickle_in)              #pickling end........
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

```python
predict_data = np.array([[2,1,4,3,2,1,5,4,3],[6,7,4,3,2,5,5,4,3]])
```

```python
print(clf.predict(predict_data))
```
